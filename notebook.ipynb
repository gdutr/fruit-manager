```python
inventaire = {
    "bananes" : 120,
    "mangues" : 85,
    "ananas" : 45,
    "noix de coco" : 60,
    "papaye" : 30
}
```

```python
def afficher_inventaire(inventaire:dict[str, int]):
    print("Inventaire actuel de plantation :")
    for fruit, quantite in inventaire.items():
        print(f"- {fruit.capitalize()} : {quantite} unités" )
```

```python
afficher_inventaire(inventaire)
```
